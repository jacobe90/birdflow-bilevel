```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# Grid size
grid_size = (20, 20)
rows, cols = grid_size
x = np.linspace(0, 1, cols)
y = np.linspace(0, 1, rows)
X, Y = np.meshgrid(x, y)

# Gaussian function
def gaussian_2d(x, y, mu_x, mu_y, sigma):
    return np.exp(-((x - mu_x)**2 + (y - mu_y)**2) / (2 * sigma**2))

# Gaussian parameters
sigma = 0.08
mu_red_1 = (0.2, 0.2)
mu_green_1 = (0.8, 0.8)
mu_red_2 = (0.4, 0.4)
mu_green_2 = (0.8, 0.8)

# Compute and normalize Gaussians
def compute_normalized_gaussian(mu_red, mu_green):
    red_gaussian = gaussian_2d(X, Y, mu_red[0], mu_red[1], sigma)
    green_gaussian = gaussian_2d(X, Y, mu_green[0], mu_green[1], sigma)
    red_gaussian /= red_gaussian.sum()
    green_gaussian /= green_gaussian.sum()
    return red_gaussian - green_gaussian

# Create the figure and axes, set the figsize to ensure plots are side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))  # Adjusting figsize to fit two plots side by side

# Plot 1
im1 = ax1.imshow(compute_normalized_gaussian(mu_red_1, mu_green_1), cmap=mpl.colormaps['bwr'])
ax1.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
ax1.text(3, 8.5, "$\mu_{t}$", fontsize=20, color='black')
ax1.text(14, 11, "$\hat{\mu}_{t}$", fontsize=20, color='black')

# Plot 2
im2 = ax2.imshow(compute_normalized_gaussian(mu_red_2, mu_green_2), cmap=mpl.colormaps['bwr'])
ax2.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
ax2.text(3, 8.5, "$\mu_{t}$", fontsize=20, color='black')
ax2.text(14, 11, "$\hat{\mu}_{t}$", fontsize=20, color='black')

# Create a single colorbar that spans both axes and place it on the right side
cbar = fig.colorbar(im1, ax=[ax1, ax2], orientation='vertical')

# Adjust layout to avoid overlap
# plt.tight_layout()

# Display the plot
plt.show()
```
